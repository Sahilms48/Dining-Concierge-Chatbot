```python
import sqlite3
import json

# Connect to the SQLite database
conn = sqlite3.connect('restaurants.db')
cursor = conn.cursor()

# Query the required fields
cursor.execute("SELECT id, cuisine FROM restaurants")
rows = cursor.fetchall()

# Create a bulk JSON file for OpenSearch ingestion
with open('bulk_restaurants.json', 'w') as f:
    for row in rows:
        restaurant_id, cuisine = row
        # Write index metadata for each document
        f.write(json.dumps({ "index": { "_index": "restaurants", "_id": restaurant_id } }) + '\n')
        # Write the document data
        f.write(json.dumps({ "restaurant_id": restaurant_id, "cuisine": cuisine }) + '\n')

print("bulk_restaurants.json generated successfully.")
conn.close()
```
